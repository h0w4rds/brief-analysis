# **Six Sigma**

[Six Sigma](https://asq.org/quality-resources/six-sigma) is a method that provides organizations tools to improve the capability of their business processes.

This increase in performance and decrease in process variation helps lead to defect reduction and improvement in profits, employee morale, and quality of products or services.

**Philosophy:** The philosophical perspective of Six Sigma views all work as processes that can be defined, measured, analyzed, improved, and controlled. Processes require inputs (x) and produce outputs (y). If you control the inputs, you will control the outputs. This is generally expressed as y = f(x).

**Set of tools:** The Six Sigma expert uses qualitative and quantitative techniques or tools to drive process improvement. Such tools include *
*statistical process control (SPC)
* control charts
* failure mode and effects analysis (FMEA)
* process mapping
* design of experiments

**Methodology**: This view of Six Sigma recognizes the underlying and rigorous DMAIC approach. 

* D=Define
* M=Measure
* A=Analyze
* I=Improve
* C=Control

While DMAIC is not the only Six Sigma methodology in use, it is certainly the most widely adopted and recognized.

**Metrics**: In simple terms, Six Sigma quality performance means 3.4 defects per million opportunities (accounting for a 1.5-sigma shift in the mean).

## **Common Cause and Assignable Cause Variation**

In many processes, regardless of how well-designed they are or carefully maintained, a certain amount of inherent or natural variability will always exist.  We refer to this natural or background or random variability as **Common Cause Variation**.
 
A process that is operation with only common cause variation present is said to be operating in statistical control or **in control**.
 
Other kinds of variability may occasionally be present in the process. For example in a Coca-cola plant a new operator could inject too much carbon dioxide into a batch.  Or a measurement can be wrong.  Or the temperature might be off, etc.  
 
This type of variation is typically large when compared to the background variation and typically represents an unacceptable level of process performance.  We refer to these sources of variation as **Assignable Cause Variation.**  A process that is operating the presence of assignable cause variation is said to be **out of control**.
 
**A major objective of process monitoring is to quickly detect when a process has gone from in control to out of control.**




## **Control Charts**

Below is an image of a typcial control chart.

image.png

A typical control chart plots the quality characteristic over time.  
 
* Center line (CL) = average value of the quality characteristic
* Upper Control Limit (UCL)
* Lower Control Limit (LCL) 
 
The UCL and LCL are chosen such that if the process is in control nearly all of the sample points will fall between the control limits.
 
There are multiple ways a control chart signals out of control behavior:
 
* When a point plots outside of the control limits that is evidence that the process is out of control.
* If the points behave in a non-random manner

You can think of a control chart as a test of the hypothesis that the process is in a state of statistical control. 

The general model for a control chart:
 
Let W be a sample statistic that measures some quality characteristic of interest, and suppose that the mean of W is $\mu_W$ and the standard deviation of W is $\sigma_w$.  The the centerline and the upper and lower control limits become:


$$\text{UCL}=\mu_w+k\sigma_w $$
$$CL=\mu_w$$
$$\text{LCL}=\mu_w-k\sigma_w $$

Where $k$ is the "distance" of the control limits away from the center line, typically expressed in standard deviation units.

## **Desiging A Control Chart: $\bar{X}$ and R Charts**

Turn around times for complete blood counts (CBC) are a critical metric within a hospital. The manager of the group has separated urgent requests from routine requests.  Everyday a random sample of five turnaround times is taken. The data is contained in the file "CBC.csv".

```r
df<-read.csv("sample_data/CBC.csv")
df
```

To plot this data we are going to take the average turnaround time for each day.

```r
df$mean <- rowMeans(df[,2:6])
df
```

```r
install.packages("qcc") #this doesn't take very long
```

```r
library(qcc)
library(tidyverse)
```

We will plot that statistic on what is called an **Xbar chart**.  This chart will capture the **day to day variability** in the turnaround time.  
 
 In general Xbar charts capture the sample to sample variability. 


```r
qcc(df[,2:6], type="xbar") # plot samples 1-5, exclding Day and sample means
```

Just for fun, check the mean of the means.

```r
mean(df$mean)
```

The control limits, UCL and LCL were calculated based on the sample ranges.  We use sample ranges when the subgroup size (samples per day in this case) is small and the number of samples per day is constant.

```r
for (i in 1:nrow(df)) {
  df$range[i] <- max(df[i, 2:6]) - min(df[i, 2:6])
}
df
```

```r
qcc(df[,2:6], type="R")
```

The Range chart or R chart captures the **within subgroup** variation.  Or in this case the within Day variation.  The center line is the average Range, $\bar{R}$.  And the control limits are based on a multiple of $\bar{R}$.

You probably already know the theory behind $\bar{X}$ and R charts.  We are plotting sample averages.  For small samples if we assume the individual observations are mostly normal (mainly symmetric) then then we know the sampling distribution of the sample mean will be normal with mean $\mu$ and $\sigma_{\bar{X}}=\frac{\sigma}{\sqrt{n}}$.

We don't know $\mu$ so we estimate it with the overall sample average and we don't know $\sigma_{\bar{X}}$ so we estimate it, in this case, with $\bar{R}$.

These charts are a pair.  
 
* We first look to make sure the within sample variation is stable, affected by only common cause variation.  In other words we see if the $\bar{R}$ chart is in-control.
* If so then we can reasonably estimate the control limits for the $\bar{X}$ chart and then we assess if the day to day variation is in control. 

## **Assessing Out of Control Behavior in the Charts**

* If both charts show no points outside the limits and no patterns, the process is said to be operating in control.  The process is only subject to its common cause variation, it is stable and predictable. 
* If the R chart shows a point outside the control limits, stop. The within subgroup variation is not in control and the limits on the $\bar{X}$ chart will not be valid.
* If the R chart is in control then we assess the $\bar{X}$ chart.  If the $\bar{X}$ chart shows a point out or another pattern the process is said to be out of control and has assignable causes present.

# **How poeple use Control Charts**

* Phase I: Retrospectively use data to establish the operating control limits.  Estimate process parameters.
* Phase II: Monitor new observations to assess process stability.
* Used to estimate current process parameters.
* Used to find assignable causes in a process.

# **Control Charts for Other Situations**

We have covered the most basic control chart for continuous data.  Aside from the $\bar{X}$ and R chart there is there are time weighted control charts:
 
* Cumulative Sum Chart
* Exponentially Weighted Moving Average Chart (EWMA)
 
The EWMA chart is non-parametric, can be used for individual observations or subgroups and can be tailored to a specific sensitivity.
 
For categorical data (attribute data as its called in the process control literature) there are also charts:
 
* $p$ chart or proportion chart
* $c$ chart or $u$ chart for counts
